```python
import torch 
from einops import rearrange, einsum
```

```python
images = torch.randn(64, 128, 128, 3) # (batch, height, width, channel) 
dim_by = torch.linspace(start=0.0, end=1.0, steps=10)  ## Reshape and multiply  
print(images.shape, dim_by.shape)
print(dim_by)
```

```python
dim_value = rearrange(dim_by, "dim_value -> 1 dim_value 1 1 1") 
print(dim_value.shape)
images_rearr = rearrange(images, "a b c d -> a 1 b c d") 
print(images_rearr.shape)
dimmed_images = images_rearr * dim_value  
print(dimmed_images.shape)
```

```python
## Or in one go:  
dimmed_images2 = einsum( images, dim_by,
                        "batch height width channel, dim_value -> batch dim_value height width channel" )
```

```python
print(dimmed_images2.shape)
```

```python

def bytes_to_unicode():
    """
    Returns a dictionary mapping byte values (0-255) to unicode strings.
    This is used to ensure a reversible mapping between bytes and unicode for BPE.
    """
    bs = list(range(33, 127)) + list(range(161, 173)) + list(range(174, 256))
    cs = bs[:]
    print("Initial byte values:", bs)
    print("Initial unicode values:", cs)
    n = 0
    for b in range(256):
        if b not in bs:
            bs.append(b)
            cs.append(256 + n)
            n += 1
    print("Extended byte values:", bs)
    print("Extended unicode values:", cs)        
    byte_to_unicode = dict(zip(bs, (chr(c) for c in cs)))
    print("Final byte to unicode mapping:", byte_to_unicode)
    return byte_to_unicode

byte_to_unicode = bytes_to_unicode()
```
